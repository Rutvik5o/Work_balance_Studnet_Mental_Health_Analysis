```python
#Series Question
habits = df["Dietary_Habits"].unique()
habits
#series-> 1D labelled array
```

```python
#we have multiple habits so we can do with for loop for finding mean
```

```python
avg_of_satisfaction = {}
for habit in habits:
    each_habit_data = df[df["Dietary_Habits"] == habit]

    avg = each_habit_data["Study_Satisfaction"].mean()

    avg_of_satisfaction[habit] = round(avg,2)

print(avg_of_satisfaction)

```

```python
#doing series

avg_series_data = pd.Series(avg_of_satisfaction)
print(avg_series_data)
```

```python
print("Habit has highest satisfaction-> Others:",avg_series_data["Others"])
```

```python
# | Task                    | Using `loc`                | Using `iloc`          |
# | ----------------------- | -------------------------- | --------------------- |
# | Get row by **label**    | `df.loc["x"]`              | ❌                     |
# | Get row by **position** | ❌                          | `df.iloc[0]`          |
# | Get range (inclusive)   | `df.loc["x":"z"]`          | `df.iloc[0:3]`        |
# | Filter condition        | `df.loc[df["Marks"] > 90]` | ✅ works only with loc |
```
